### MinMaxScaler:
 

```python
import wrangle_zillow as w 
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = w.aquire_zillow_data()
df = w.prepare(df)
train, validate, test = w.split_data(w.prepare(w.aquire_zillow_data()))
```

```python
train = train.sample(3000)
validate = validate.sample(3000)
test = validate.sample(3000)
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
train_scaled = scaler.transform(train)
validate_scaled = scaler.transform(validate)
test_scaled = scaler.transform(test)
```

```python
plt.figure(figsize=(13, 6))
plt.subplot(121)
plt.hist(train, bins=25, ec='black')
plt.title('Original')
plt.subplot(122)
plt.hist(train_scaled, bins=25, ec='black')
plt.title('Scaled')
```

### StandardScaler:

```python
standard_scaler = StandardScaler()
```

```python
standard_scaler.fit(train)
train_scaled = standard_scaler.transform(train)
validate_scaled = standard_scaler.transform(train)
test_scaler = standard_scaler.transform(test)

plt.figure(figsize=(13,6))
plt.subplot(121)
plt.hist(train,bins = 25, ec='red')
plt.title('Original')
plt.figure(figsize=(13,6))
plt.subplot(122)
plt.hist(train,bins = 25, ec='red')
plt.title('Scaled')
```

### RobustScaler

```python
robust_scaler = RobustScaler()

robust_scaler.fit(train)
train_scaled =  robust_scaler.transform(train)
validate_scaled = robust_scaler.transform(validate)
test_scaled = robust_scaler.transform(test)

plt.figure(figsize=(13, 6))
plt.subplot(121)
plt.hist(train, bins=25, ec='black')
plt.title('Original')
plt.subplot(122)
plt.hist(train_scaled, bins=25, ec='black')
plt.title('Scaled')

```
